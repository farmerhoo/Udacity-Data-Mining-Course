# Google Apps 商店的数据分析

## 简介

相信大家对移动应用商店都不陌生吧。Google Play Store（Google Play商店）是谷歌官方的软件应用商店，拥有上架软件数十万款，下载量更是突破了20亿次，为了手机用户提供了极为广泛的应用选择，很受大家的欢迎。



本数据集(googleplaystore.csv)包含了 Google Play 商店中 App 的数据。该数据是Kaggle中 [Google Play Store Apps](https://www.kaggle.com/lava18/google-play-store-apps) 的一部分，其中包含 Google Play 商店中 10k+ 应用软件的信息。

数据中的变量含义解释：
```
App: 应用的名称，字符变量。
Category: 应用所属的分类，字符变量。
Rating: 某应用的用户评分，数值变量。
Reviews: 某应用获得的用户评论数量，数值变量。
Size: 某应用的所占存储空间的大小，字符变量。
Installs: 用户安装和下载某应用的次数，字符变量。
Type: 付费或免费，分类变量。
Price: 价格，字符变量。
Content Rating: 应用商店针对内容给出的年龄评级组 - Children / Mature 21+ / Adult，分类变量。
Genres: 类型/流派，一个应用可以属于多个流派，比如音乐、游戏、家庭等，字符变量。
Last Updated: 应用最新更新的日期，字符变量。
Current Ver: 当前应用的版本，字符变量。
Android Ver: 安装该应用所需要的最低安卓版本，字符变量。

```

**查看Reviews的统计特性**

````python
# 将Reviews列的数据数值化
import re # 导入正则表达式

reviews = np.array(df['Reviews'])

# 创建array存放数据化后的
reviews_num = np.array([]) 

for review in reviews:
    if re.match("[0-9]*\.?[0-9]+M$", review):
        num = int(float(review[:-1]) * 1e6)
        reviews_num = np.append(reviews_num, num)
    elif re.match("[0-9]*\.?[0-9]+K$", review):
        num = int(float(review[:-1]) * 1e3)
        revirews_num = np.append(reviews_num, num)
    else:
        num = int(review)
        reviews_num = np.append(reviews_num, num)
````

````python
# 查看Reviews列的统计特性
df['reviews_num'] = reviews_num
df.describe()
````

可以看出，评论数量最大再7815万，最少评论量为0。

平均值评论量和中位数相差较远，且中位数远小于平均数，说明可能存在部分APP下载量较大而拉高整体评论量的情况，数据的方差较大也印证了这一情况。

**查看Installs的统计特性**

````python
# 将Installs列的数据数值化

installs = np.array(df['Installs'])

# 创建array存放数据化后的
installs_num = np.array([]) 

for install in installs:
    if re.match("[0-9][0-9,]+\+$", install):
        install = install.replace(',', '')
        num = int(install[:-1])
        installs_num = np.append(installs_num, num)
    elif re.match("[0-9]+\+$", install):
        num = int(install[:-1])
        installs_num = np.append(installs_num, num)        
    elif install == 'Free':
        num = 0
        installs_num = np.append(installs_num, num)
    else:
        num = int(install)
        installs_num = np.append(installs_num, num)
````

````python
# 查看Installs列的统计特性
df['installs_num'] = installs_num
df.describe()
````

根据下载量Installs的统计数据可以发现：

1. 最大下载量为1亿
2. 最小下载量为0
3. 平均下载量为1500多万
4. 下载量中位数为10万

**总结**：

最大下载量和最小下载量悬殊很大，而且中位数远小于平均数，有强者越强的特点，应该是大下载量拉高了整体的平均值，但是实际上APP的下载量中位数并不高，能做到10万下载量的APP

**查看Price统计特性**

````python
# 将Price列的数据数值化

prices = np.array(df['Price'])

# 创建array存放数据化后的
prices_num = np.array([]) 

for price in prices:
    if price.startswith('$'):
        num = float(price[1:])
        prices_num = np.append(prices_num, num)       
    elif price == '0' or price == 'Everyone':
        num = 0
        prices_num = np.append(prices_num, num)
    else:
        num = float(price)
        prices_num = np.append(prices_num, num)
````

````python
# 查看Price列的统计特性
df['Price_num'] = prices_num
df.describe()
````

## 3 数据探索分析

### 3.1 分析Rating评分数据

````python
# 取出rating不缺失的数据
df_rating = df[df['Rating'].isnull()==False]
df_rating.head()
````

````python
# rating列数据进行排序
df_rating.sort_values(by='Rating', ascending=False, inplace=True)
df_rating.head(10)
````

这里发现Rating评分前10的都是满分5分，考虑到评分数量reviews较少的APP的评分可能不够客观，接下来对评分Rating列和Reviews列共同排序。

评分和下载量最高的APP是Ríos de Fe

````python
# 将df_rating分为付费类df_rpaid和免费类df_rfree
df_rpaid = df_rating[df['Type']=='Paid']
df_rfree = df_rating[df['Type']=='Free']
print('付费类APP评分从高到低排序TOP10:')
print(df_rpaid['App'][0:10])
print('免费类APP评分从高到低排序TOP10:')
print(df_rfree['App'][0:10])
````

最高评分的付费类APP：RETRO Shocked DW-6000
