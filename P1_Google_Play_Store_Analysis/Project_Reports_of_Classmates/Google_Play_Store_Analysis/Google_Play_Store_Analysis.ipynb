# Google Apps 商店的数据分析

## 简介

相信大家对移动应用商店都不陌生吧。Google Play Store（Google Play商店）是谷歌官方的软件应用商店，拥有上架软件数十万款，下载量更是突破了20亿次，为了手机用户提供了极为广泛的应用选择，很受大家的欢迎。



本数据集(googleplaystore.csv)包含了 Google Play 商店中 App 的数据。该数据是Kaggle中 [Google Play Store Apps](https://www.kaggle.com/lava18/google-play-store-apps) 的一部分，其中包含 Google Play 商店中 10k+ 应用软件的信息。

数据中的变量含义解释：
```
App: 应用的名称，字符变量。
Category: 应用所属的分类，字符变量。
Rating: 某应用的用户评分，数值变量。
Reviews: 某应用获得的用户评论数量，数值变量。
Size: 某应用的所占存储空间的大小，字符变量。
Installs: 用户安装和下载某应用的次数，字符变量。
Type: 付费或免费，分类变量。
Price: 价格，字符变量。
Content Rating: 应用商店针对内容给出的年龄评级组 - Children / Mature 21+ / Adult，分类变量。
Genres: 类型/流派，一个应用可以属于多个流派，比如音乐、游戏、家庭等，字符变量。
Last Updated: 应用最新更新的日期，字符变量。
Current Ver: 当前应用的版本，字符变量。
Android Ver: 安装该应用所需要的最低安卓版本，字符变量。

```

## 数据探索分析

````python
# In exploratory data analysis, please make sure of using statistics and visualizations
#查看安装量排名前10的App
dfs_install = dfs.sort_values(by='Installs',ascending=False)
dfs_install.head(10)
````

通过观察发现:10条数据安装量都超过1亿人次，但是很有意思的是，在数据中发现了重复项，例如Google，Instagram，重复项中仅有评论数略有差别，其他均相同，猜测可能是相近时间点数据合并更新时未能去除

````python
#首先查看还有多少重复数据
#选取App和Current Ver以及收费类型三项参数来定位，版本相同，名称和收费类型相同，应视为同一App
sum(dfs.duplicated(subset=['App','Current Ver','Type']))
````

````python
#671条，不容小视，drop之
#虽然评论数差别不大，我们还是按评论数排序，留下较多的那个
dfs.sort_values(by='Reviews',ascending=False,inplace=True)
dfs.drop_duplicates(subset=['App','Current Ver','Type'],keep='first',inplace=True)
dfs.info()
````

````python
#再对新数据按照安装量、评论数、评分三项指标进行排序
dfs.sort_values(by=['Installs','Reviews','Rating'],ascending=False,inplace=True)
#重置索引
dfs.reset_index(drop=True,inplace=True)
dfs.head(10)
````

安装量排前10名的都是免费App，尝试按照收费类型分组进行对比

````python
#按照收费类型分组
dfs_free = dfs[dfs['Type']=='Free'].sort_values(by=['Installs','Reviews','Rating'],ascending=False)
dfs_paid = dfs[dfs['Type']=='Paid'].sort_values(by=['Installs','Reviews','Rating'],ascending=False)
````

````python
#比较付费和免费App的数量
fig = plt.figure(figsize=(12,6))
ax1 = fig.add_subplot(1,2,1)
ax2 = fig.add_subplot(1,2,2)

#直方图展示
num = [dfs_free['App'].count(),dfs_paid['App'].count()]
name = ['Free','Paid']
ax1.set_title('App数量直方图对比',fontsize=24)
ax1.bar(name,num,color=['r','g'])
ax1.tick_params(axis='both',which='major',labelsize=14)

#饼图对比
ax2.set_title('App数量饼图对比',fontsize=24)
labels = ['Free','Paid']
explode = (0,0.1)
sizes = [
    100*dfs_free['App'].count()/(dfs_free['App'].count()+dfs_paid['App'].count()),
    100*dfs_paid['App'].count()/(dfs_free['App'].count()+dfs_paid['App'].count())
        ]
ax2.pie(sizes,explode=explode,labels=labels, autopct='%1.1f%%',
        shadow=True, startangle=90,colors=['r','g'])
ax2.axis('equal')


plt.show()
````

发现：免费App牢牢占据主导地位

````python
#继续比较收费App和免费App的平均评论数量和平均评分
fig = plt.figure(figsize=(12,10))
ax1 = fig.add_subplot(2,2,1)
ax2 = fig.add_subplot(2,2,2)
ax3 = fig.add_subplot(2,2,3)
ax4 = fig.add_subplot(2,2,4)


#先比较两种类型的平均评论数
name = ['Free','Paid']
review = [dfs_free['Reviews'].mean(),dfs_paid['Reviews'].mean()]
ax1.bar(name,review,color=['r','g'])
ax1.set_title('平均评论数量对比',fontsize=24)

#比较两种类型的平均评分
rating = [dfs_free['Rating'].mean(),dfs_paid['Rating'].mean()]
ax2.bar(name,rating,color=['r','g'])
ax2.set_title('平均评分对比',fontsize=24)

#比较两种类型的平均下载量
install = [dfs_free['Installs'].mean(),dfs_paid['Installs'].mean()]
ax3.bar(name,install,color=['r','g'])
ax3.set_title('平均下载量对比',fontsize=24)

#比较两种类型的平均下载评论比
scale = [dfs_free['Reviews'].mean()/dfs_free['Installs'].mean(),
         dfs_paid['Reviews'].mean()/dfs_paid['Installs'].mean()]
ax4.bar(name,scale,color=['r','g'])
ax4.set_title('平均下载评论比对比',fontsize=24)
````

可以看到：在平均下载量和评论数上免费App都碾压付费App，但是在平均评分付费App略高于免费App（付费App的平均质量略好），而且付费用户更倾向于对App发表评论，大概是免费用户的4倍

````python
#下面对App的类别进行研究，看看哪些类别的App更受欢迎
#观察各类App的总下载量和平均下载量
sum_category = dfs['Installs'].groupby(dfs['Category']).sum().sort_values(ascending=False)
mean_category = dfs['Installs'].groupby(dfs['Category']).mean().sort_values(ascending=False)
fig,axes = plt.subplots(1,2)
sum_category[:10].plot(kind='bar',ax=axes[0],title='App下载总量前十名',figsize=(16,6),rot=45);
mean_category[:10].plot(kind='bar',ax=axes[1],title='App平均下载量前十名',rot=45);
````

可以看出总下载量方面：游戏、通讯、工具领先，说明对这几类App更受欢迎；平均下载量方面：通讯、社交、录像领先，说明在这几类App的数量相对其他总下载量大的App较少。

````python
#推测：安装数量更多的App的评论数会更多，观察下是否有这样的相关性
#按照下载量分组，查看其平均评论数
mean_review = dfs['Reviews'].groupby(dfs['Installs']).mean().sort_values(ascending=True)
plt.figure(figsize=(10,6))
#由于数量级比较大，且比较离散，所以取10为底的对数处理
plt.scatter(x=np.log10(mean_review.index),y=np.log10(mean_review),color='g')
plt.title('App用户下载量与平均平均数关系',fontsize=24)
plt.xlabel('用户下载量',fontsize=14)
plt.ylabel('平均评论数',fontsize=14)

plt.show()
````

发现：基本符合推测，即用户下载量和App获取的平均评论数呈正相关，因为用户下载量数据是离散的，且分组比较少，所以得到数据并不是很多。

````python
#继续看一下不同的用户下载量分组下App的平均评分会不会呈现相似的相关性
mean_rating = dfs['Rating'].groupby(dfs['Installs']).mean().sort_values(ascending=True)
plt.figure(figsize=(10,6))
#评分数据最高5分，而下载量最高达10亿次，所以取对数处理
plt.scatter(x=np.log10(mean_rating.index),y=mean_rating,color='g')
plt.title('App用户下载量与平均评分关系',fontsize=24)
plt.xlabel('下载量',fontsize=14)
plt.ylabel('平均评分',fontsize=14)

plt.show()
````

发现：下载量在10000次前后，数据具有不同的趋势。10000次以下时，下载量10为底的对数与平均评分呈负相关，超过10000次后，呈正相关

````python
#探索用户评论数与评分之间的关系
plt.figure(figsize=(10,6))
plt.scatter(x=np.log10(dfs['Reviews']),y=dfs['Rating'],color='g')
plt.title('App用户评论数与评分关系散点图',fontsize=24)
plt.xlabel('用户评论数（取对数）',fontsize=14)
plt.ylabel('评分',fontsize=14)

plt.show()
````

发现存在比较弱的正相关性，继续分组探索
